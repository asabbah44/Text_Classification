```
#list of libraries used by the code
import string
from gensim.models import Word2Vec
import logging
from nltk.corpus import stopwords
from textblob import Word
import pandas as pd
import nltk
import re
import spacy
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
stop_words = nltk.corpus.stopwords.words('english')
df1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Data/Alldata.csv")
df2 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Data/messages.csv")
df3=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Data/MyAllComents.csv")

df1=pd.DataFrame(df1, columns = ['commenttext'])
df2=pd.DataFrame(df2, columns = ['message'])
df3=pd.DataFrame(df3, columns = ['CommentsText'])

df1['commenttext']=df1
df2['commenttext']=df2
df3['commenttext']=df3
df=pd.concat([df1,df2,df3])

def preprocess_text(sen):
    # Removing html tags
    sentence = remove_tags(sen)

    # Remove punctuations and numbers
    sentence = re.sub('[^a-zA-Z]', ' ', sentence)

    # Single character removal
    sentence = re.sub(r"\s+[a-zA-Z]\s+", ' ', sentence)

    # Removing multiple spaces
    sentence = re.sub(r'\s+', ' ', sentence)

    return sentence

TAG_RE = re.compile(r'<[^>]+>')

def remove_tags(text):
    return TAG_RE.sub('', text)


# Cleaing the text
import sys
sys.path.append('/content/drive/MyDrive/Colab Notebooks')

from Utility import preprocess
df["commenttext"] = df["commenttext"].apply(lambda x: x.lower())

# Clean the text
df["commenttext"] = df["commenttext"].apply(lambda x:preprocess.clean_text(x))

# Clean numbers
df["commenttext"] = df["commenttext"].apply(lambda x:preprocess.clean_numbers(x))

# Clean Contractions
df["commenttext"] = df["commenttext"].apply(lambda x:preprocess.replace_contractions(x))
 
df["commenttext"] = df["commenttext"].apply(lambda x:preprocess.clean_punc(x))

# lemmatize
df["commenttext"] = df["commenttext"].apply(lambda x: preprocess.lemmatize(x))
# remove stop word and custome word such as todo
#df["commenttext"] = df["commenttext"].apply(lambda x: preprocess.custm_stopwords_remove(x))

X = []
sentences = list(df['commenttext'])
for sen in sentences:
    X.append(preprocess_text(sen))

# Preparing the dataset

all_sentences =X# nltk.sent_tokenize(X)

all_words = [nltk.word_tokenize(sent) for sent in all_sentences]

# Removing Stop Words
from nltk.corpus import stopwords
new_words=('http','https','xxx','todo','camel','li','gerrit','wordpres','log4j','tomcat','k9','hadoop')
for word in  new_words:
    stop_words.append(word)
for i in range(len(all_words)):
    all_words[i] = [w for w in all_words[i] if w not in stop_words]

model = Word2Vec(all_words, min_count=2,size=100,workers=4,window=3)

words = len(list(model.wv.vocab))
print(words)
print(list(model.wv.vocab))
```
